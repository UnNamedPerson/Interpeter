```python

# token class to define the token, which is an character that is read from the input and has a type and value

class token():
    def __init__(self, value, type):
        self.value = value # value is any thiscase any number form from zero to nine 
        self.type = type # type here can be a number (Integer), operation (plus sign in the case of our class), or EOF (end of file).

print("hello")
```

```python
# a scanner class to read each character and analyze it. the class identifies whehter or the token is
# following the right rules of the add experssion. 
import math
class scanner():
    def __init__(self, text):
        self.text = text
        self.currentToken = None
        self.pos = 0
        
#   this function gets the next token in the text that is passed by the user and returns a new token based on the character
#   that was read.
    def getNextToken(self):
        if self.pos == len(self.text):
            return token(None, "EOF")    
        
#       if we are not at the end of the file, we choose the current character as the value for our token.
        currentChar = self.text[self.pos]

        if currentChar.isdigit():
            self.pos += 1
            return token(int(currentChar), "Integer")
        
        elif currentChar == "+":
            self.pos +=1
            return token(currentChar, "Plus")
        
        elif currentChar == "-":
            self.pos +=1
            return token(currentChar, "Minus")
        
        elif currentChar == "*":
            self.pos +=1
            return token(currentChar, "Mult")
        
        elif currentChar == "/":
            self.pos +=1
            return token(currentChar, "Div")
        
        elif currentChar == " ":
            self.pos +=1
            return token(currentChar, "space")
    
    
#   a function that handels white space and consumes it.  
    def consumeWhiteSpace(self):
        while self.currentToken.value == " ":
            self.consumeToken(self.currentToken.type, "space")
            self.currentToken = self.getNextToken()
        
    # this function is built to handle different operations found between integers.
    def consumeOpToken(self):
        if self.currentToken.value == "-":
            self.consumeToken(self.currentToken.type, "Minus")
            operation = "-"
            
        elif self.currentToken.value == "+":
            self.consumeToken(self.currentToken.type, "Plus")
            operation = "+"
            
        elif self.currentToken.value == "*":
            self.consumeToken(self.currentToken.type, "Mult")
            operation = "*"
            
        elif self.currentToken.value == "/":
            self.consumeToken(self.currentToken.type, "Div")
            operation = "/"
        
        self.currentToken = self.getNextToken()   
        return operation

#   this function turns a list of digits into one int to be returned as one value, so (1,3,4) = (134)
    def convertToInt(self, intList):
        return sum(digit * 10 ** i for i, digit in enumerate(intList[::-1]))
        
    #    this function is to check if the current token matches the token's type in the right order.
    #    If it is in the right order, returns the current token, if not raise error.
    def consumeToken(self,currentTokenType, correctType):
        if currentTokenType == correctType:
            return self.currentToken
        else: 
            raise Exception("error while parsing the input") #raise an error if we encounter an error while parsing the token.
                      

    def term(self):
        self.currentToken = self.getNextToken()
        leftIntList = list()
#       consuming the first number  
        while(isinstance(self.currentToken.value, int)) :
#             print("inside while: ", self.currentToken.value)
            self.consumeToken(self.currentToken.type, "Integer")
            leftIntList.append(self.currentToken.value)
            self.currentToken = self.getNextToken()
#         print("list: ", leftIntList)
        leftInt = self.convertToInt(leftIntList)
        
        if self.currentToken.type != "EOF":        
    #       consuming white space between the number and the operation  
            self.consumeWhiteSpace()
            operation = self.consumeOpToken()
    #       consuming white space between the operation and the other number 
#             self.consumeWhiteSpace()
                
            if operation == "+":
                print("leftInt is: ", leftInt)
                result = leftInt + self.term()
                return result
            elif operation == "-" :
                
                print("leftInt is: ", leftInt)
                result = leftInt - self.term()
                return result
            elif operation == "*":
                result = leftInt * self.term()
                return result
            else:
                result = leftInt / self.term()
                return result
        else:
#             print("number: ", leftInt)
            return leftInt
                

#   this function checks if the passed text stisfies the required format.        
    def Expr(self):
        return self.term()
#     apparetnly 
        
```
